```python
import re
```

```python
test = "1D2S#10S"
```

```python
def solution(dart_result: str) -> int:
    character = re.compile("[A-Z]+")
    digit = re.compile("[0-9]{1,2}")
    option = re.compile("\W")

    all_character = character.findall(dart_result)
    all_digit = digit.findall(dart_result)
    all_option = option.findall(dart_result)

    cha2digit = {'S': 1, 'D': 2, 'T': 3}
    all_character = [cha2digit[cha] for cha in all_character]

    term_1 = int(all_digit[0]) ** all_character[0]
    term_2 = int(all_digit[1]) ** all_character[1]
    term_3 = int(all_digit[2]) ** all_character[2]
    oops_1 = 1
    oops_2 = 1
    oops_3 = 1
    star_1 = 0
    star_2 = 0
    star_3 = 0

    if len(all_option) == 1:
        idx = dart_result.index(all_option[0])
        if dart_result[idx] == '#':
            if idx < 4:
                oops_1 = -1
            elif idx >= 4 and idx < 6:
                oops_2 = -1
            elif idx >= 6:
                oops_3 = -1
        else:
            if idx < 4:
                star_1 += 1
            elif idx >= 4 and idx < 6:
                star_1 += 1
                star_2 += 1
            elif idx >= 6:
                star_2 += 1
                star_3 += 1

    elif len(all_option) == 2:
        p = re.compile('\W')
        m = p.finditer(dart_result)
        option_idxs = [r.start() for r in m]

        for a in m:
            idx = a.start()        
            if idx < 4:
                if dart_result[idx] == '#':
                    oops_1 = -1
                else:
                    star_1 += 1
            elif idx >= 4 and idx < 7:
                if dart_result[idx] == '#':
                    oops_2 = -1
                else:
                    star_1 += 1
                    star_2 += 1
            elif idx >= 7:
                if dart_result[idx] == '#':
                    oops_3 = -1
                else:
                    star_2 += 1
                    star_3 += 1

    elif len(all_option) == 3:

        p = re.compile('\W')
        m = p.finditer(dart_result)
        option_idxs = [r.start() for r in m]

        for a in m:
            idx = a.start()

            if idx < 4:
                if dart_result[idx] == '#':
                    oops_1 = -1
                else:
                    star_1 += 1
            elif idx > 4 and idx < 8:
                if dart_result[idx] == '#':
                    oops_2 = -1
                else:
                    star_1 += 1
                    star_2 += 1
            elif idx >= 8:
                if dart_result[idx] == '#':
                    oops_3 = -1
                else:
                    star_2 += 1
                    star_3 += 1

    result = term_1 * (oops_1) * (2 ** star_1) +\
             term_2 * (oops_2) * (2 ** star_2) +\
             term_3 * (oops_3) * (2 ** star_3)
    
    return result
```
